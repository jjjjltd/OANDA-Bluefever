```python
import requests
import defs
import pandas as pd
```

Open a session so that requests don't have to re-connect each time

```python
session = requests.Session()
```

Consider turning instrument(s) into a list...

```python
instrument = "EUR_USD"
```

```python
params = dict(
    count = 10,
    granularity = "H1",
    price="MBA"
)
params
```

```python
url = f"{defs.OANDA_URL}/instruments/{instrument}/candles"
url
```

```python
response = session.get(url, params=params, headers=defs.SECURE_HEADER).json()
response
```

```python
prices = ["bid", "mid", "ask"]
ohlc = ["o", "h", "l", "c"]
```

```python
our_data = []

for candle in response["candles"]:
    if candle["complete"] == False:
        continue

    new_dict = {}
    new_dict["ticker"] = response["instrument"]
    new_dict["time"] = candle["time"]
    new_dict["volume"] = candle["volume"]
    
    for price in prices:
        for oh in ohlc:
            new_dict[f"{price}_{oh}"] = candle[price][oh]

    our_data.append(new_dict)
```

```python
candles_df = pd.DataFrame.from_dict(our_data)
candles_df.to_csv("checkpoint.csv")
```
