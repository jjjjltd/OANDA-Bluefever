```python
all_trades["CROSS"] = "MA_" + all_trades.MASHORT.map(str) + "_" + all_trades.MALONG.map(str)
```

```python
def plot_line(df_plot, name):
    fig=go.Figure()
    fig.add_trace(go.Scatter(
        x=df_plot.time, 
        y=df_plot.CUM_GAIN,
        line=dict(width=1, color="#3d825f"),
        line_shape='spline',
        name=name,
        mode='lines'
        ))
    fig.update_layout(width=1000, height=400, 
        margin=dict(l=15, b=15, t=35, r=10), 
        font=dict(size=10, color="#e1e1e1"),
        paper_bgcolor = "#1e1e1e", 
        plot_bgcolor = "#1e1e1e",
        title=name
        )
    fig.update_xaxes(
        linewidth=1,
        linecolor="#3a4a54",
        showgrid=False,
        zeroline=False
        )
    fig.update_yaxes(
        linewidth=1,
        linecolor="#3a4a54",
        showgrid=False,
        zeroline=False
        )
    
    fig.show()
    
```

```python
for c in crosses:
    temp_df = all_trades[(all_trades.CROSS==c)].copy()
    temp_df = temp_df.groupby(by="time", as_index=False).sum()
    temp_df["CUM_GAIN"] = temp_df.GAIN.cumsum()
    plot_line(temp_df, c)
```
