```python
import pandas as pd
import plotly.graph_objects as go
```

```python
ma_test_res = pd.read_csv("./Data/ma_test_results.csv")
all_trades = pd.read_csv("./Data/All Trades.csv")
```

```python
all_trades.column = all_trades[['time', 'PIP_Delta', 'GAIN', 'PAIR', 'MASHORT', 'MALONG', 'DURATION']]
```

```python
all_trades.head()
```

```python
ma_test_res["CROSS"] = "MA_" + ma_test_res.mashort.map(str) + "_" + ma_test_res.malong.map(str)
```

```python
ma_test_res.head()
```

```python
df_all_gains = ma_test_res.groupby(by=["CROSS", "mashort", "malong"], as_index=False).sum()
```

```python
df_all_gains.sort_values(by="total_gain", ascending=False, inplace=True)
```

```python
df_all_gains.head()
```

```python
for cross in df_all_gains.CROSS.unique():
    df_temp = ma_test_res[ma_test_res.CROSS==cross]
    total_profit = df_temp.shape[0]
    n_good = df_temp[df_temp.total_gain>0].shape[0]
    print(f"{cross:12} {n_good:4} {(n_good/total_profit)*100:4.0f}%")
```
